```python
import time
start = time.time()
from pyspark.sql.functions import split
from pyspark.mllib.fpm import FPGrowth
from pyspark import SparkConf
from pyspark.context import SparkContext
```

```python
data = (spark.read
...     .text("groceries.csv")
...     .select(split("value", ",").alias("items")))
```

```python
data.show(truncate=False)
```

```python
from pyspark.ml.fpm import FPGrowth
```

```python
fpGrowth=FPGrowth(itemsCol="items", minSupport=0.01, minConfidence=0.2)
```

```python
model=fpGrowth.fit(data)
model.freqItemsets.show(333,False)
```

```python
model.associationRules.show(231,False)
```

```python
model.transform(data).show(5900,False)
```

```python
print(f'Time: {time.time() - start}')
```
